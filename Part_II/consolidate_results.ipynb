```python
import pandas as pd
import numpy as np
import os
import json
```

```python
rows = []
for constraints in ('one', 'two', 'all'):
    for method, path in [('codex', 'codex_translations'), ('gpt', 'nl_plans')]:
        with open(os.path.join(path, constraints, 'collated.json'), 'r') as f:
            data = json.load(f)
            for row in data:
                rows.append({
                    'id': row['id'],
                    'succeed': row['succeed'],
                    'constraints': constraints,
                    'method': method,
                    'gpt_prompt': None if method != 'gpt' else row['query'],
                    'gpt_res': None if method != 'gpt' else row['lm_plan']
                })
df = pd.DataFrame(rows)
df
```

```python
wide = df.pivot(index='id', columns=['constraints', 'method'], values='succeed')
print(wide.to_string())
```

```python
print(wide.sum().unstack(level=0)[['one', 'two', 'all']].to_string())
```
