```python
'''
This is a util notebook which combines explanations generated per condition
Adjusts the batch index across condition tpye
This creates a csv that has all explanations generated by GPT-3 (*before* humanness experiment)
'''

import os
import numpy as np 
import pandas as pd 
import random
```

```python
# load in gpt-3 generations per condition
# combine and scale batch index across conditions 
# for now, read directly from .txt files
gpt3_gen_dir = "/Users/kcollins/language_and_structure_of_thoughts/explanations/gpt3/generation/gpt3_generations/"

conditions = ["unconstrained", "single-constraint", "all-constraints"]
num_batches = 20 # per condition
dfs = []

for cond_idx, cond_type in enumerate(conditions): 
    df = pd.read_csv(f"{gpt3_gen_dir}max_inductive_{cond_type}_explanations_full_batched.csv")
    df.batch_idx = [cond_idx * num_batches + b for b in df.batch_idx] # re-scale batch idx per cond
    dfs.append(df)

merged_df = pd.concat(dfs, ignore_index=True)
merged_df.to_csv(f"{gpt3_gen_dir}gpt3_generations_max_prompting_explanations_combined.csv")
```
