```python
# process plans for running the humanness experiments 

import os
import openai
from dotenv import load_dotenv
import numpy as np 
import pandas as pd 
import random

load_dotenv()

openai.api_key = "" # TODO: fill in w/ own key

random.seed(10)
```

```python
goal_types = ["unconstrained", "constrained_many", "constrained_single"] # swapped order b/c we already ran some of the single ones
save_dir = "./gpt3_generations/"
dfs = []
num_batches_per_type = 19 # NOTE: this is fixed -- adjust depending on the number of batches, or append to end like single
for idx, goal_type in enumerate(goal_types): 
    df = pd.read_csv(f"{save_dir}/max_inductive_{goal_type}_plans_full.csv")    
    # adjust batch index -- incremement by previous 
    df["batch_idx"] += (idx * num_batches_per_type)
    dfs.append(df)
merged_df = pd.concat(dfs, ignore_index=True)

# remove "Goal: " from start
for idx, goal in enumerate(merged_df.goal):
    if goal[:5] == "Goal:": 
        merged_df.goal[idx] = goal[6:]
```
